```python
def fill_country(city, country):
    if country is pd.NA:
        the_country = re.findall(r"\(\w+", city)
        if the_country != []:
            the_country = the_country[0][1:].capitalize()
            if the_country == "Uk":
                the_country = "United Kingdom"
            if the_country == "Puerto":
                the_country = "Puerto Rico"
            if the_country == "Uae":
                the_country = "United Arab Emirates"
            if the_country == "New":
                the_country = "New Zealand"
            if the_country == "İran":
                the_country == "Iran"
            if the_country == "Viet":
                the_country = "Viet Nam"
            if the_country == "Costa":
                the_country = "Costa Rica"
            if the_country == "Brasil":
                the_country = "Brazil"
            if the_country in countries:
                return the_country
        
    return country

ufos["country"] = np.vectorize(fill_country)(ufos["city"], ufos["country"])
```

## 2.5. CITY-REMOVING AFTER THE PARANTHESES

```python
ufos["city"] = pd.Series(ufos["city"], dtype="string")
# As you realise, I don't have to use coerce method because there is no nan value.
```

```python
def removing_prt(city):
    x = city.find("(")
    if x != -1:
        city = city[:x]
        return city
    return city


ufos["city"] = np.vectorize(removing_prt)(ufos["city"])
ufos["city"] = pd.Series(ufos["city"], dtype="string")
```
